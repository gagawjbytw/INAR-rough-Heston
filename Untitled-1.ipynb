```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import sys

# --- 1. Load data (same as before) ---
CSV_FILE = 'results/iv_surface_cdq.csv'

print(f"Attempting to read data from '{CSV_FILE}'...")

try:
    df = pd.read_csv(CSV_FILE, skiprows=1)
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print(f"Error: The file '{CSV_FILE}' was not found.")
    print("Please make sure you have run the './scripts/ivs.sh' script first to generate the data.")
    sys.exit(1)
except Exception as e:
    print(f"An error occurred while processing the file: {e}")
    sys.exit(1)

# --- 2. Data preparation (same as before) ---

# Get unique axis values
maturities = df['Maturity'].unique()
log_moneyness = df['LogMoneyness'].unique()

# Create the grids
T, K = np.meshgrid(maturities, log_moneyness)

# Reshape the implied volatility data to match the grid
iv_rough = df.pivot(index='LogMoneyness', columns='Maturity', values='IV_Rough').values
iv_classical = df.pivot(index='LogMoneyness', columns='Maturity', values='IV_Classical').values

print("\nData has been reshaped. Now generating plots...")

# ===================================================================
# --- Figure 1: Plot only the Rough Model (alpha < 1) surface ---
# ===================================================================

fig1 = plt.figure(figsize=(12, 9), constrained_layout=True)
ax1 = fig1.add_subplot(111, projection='3d')

surf1 = ax1.plot_surface(T, K, iv_rough * 100, cmap='viridis', rstride=1, cstride=1, antialiased=True)

ax1.set_xlabel('Maturity (T)', fontsize=12, labelpad=10)
ax1.set_ylabel('Log-Moneyness (k)', fontsize=12, labelpad=10)
ax1.set_zlabel('Implied Volatility (%)', fontsize=12, labelpad=10)
ax1.set_title('Implied Volatility Surface (Rough Model, alpha < 1)', fontsize=16, pad=20)
ax1.zaxis.set_major_formatter(FormatStrFormatter('%.1f%%'))
ax1.view_init(elev=30, azim=-120)

fig1.colorbar(surf1, shrink=0.5, aspect=10, pad=0.1, label='Rough IV (%)')
output_file_1 = 'results/iv_surface_rough.png'
fig1.savefig(output_file_1, dpi=300)
print(f"Plot 1 saved to '{output_file_1}'")


# ===================================================================
# --- Figure 2: Plot only the Classical Model (alpha = 1) surface ---
# ===================================================================

fig2 = plt.figure(figsize=(12, 9), constrained_layout=True)
ax2 = fig2.add_subplot(111, projection='3d')

surf2 = ax2.plot_surface(T, K, iv_classical * 100, cmap='plasma', rstride=1, cstride=1, antialiased=True)

ax2.set_xlabel('Maturity (T)', fontsize=12, labelpad=10)
ax2.set_ylabel('Log-Moneyness (k)', fontsize=12, labelpad=10)
ax2.set_zlabel('Implied Volatility (%)', fontsize=12, labelpad=10)
ax2.set_title('Implied Volatility Surface (Classical Model, alpha = 1)', fontsize=16, pad=20)
ax2.zaxis.set_major_formatter(FormatStrFormatter('%.1f%%'))
ax2.view_init(elev=30, azim=-120)

fig2.colorbar(surf2, shrink=0.5, aspect=10, pad=0.1, label='Classical IV (%)')
output_file_2 = 'results/iv_surface_classical.png'
fig2.savefig(output_file_2, dpi=300)
print(f"Plot 2 saved to '{output_file_2}'")


# ===================================================================
# --- Figure 3: Compare both surfaces side by side ---
# ===================================================================

fig3 = plt.figure(figsize=(14, 10), constrained_layout=True)
ax3 = fig3.add_subplot(111, projection='3d')

# Plot two semi-transparent surfaces
ax3.plot_surface(T, K, iv_rough * 100, cmap='viridis', alpha=0.7)
ax3.plot_surface(T, K, iv_classical * 100, cmap='plasma', alpha=0.7)

ax3.set_xlabel('Maturity (T)', fontsize=12, labelpad=10)
ax3.set_ylabel('Log-Moneyness (k)', fontsize=12, labelpad=10)
ax3.set_zlabel('Implied Volatility (%)', fontsize=12, labelpad=10)
ax3.set_title('Implied Volatility Surface (Comparison)', fontsize=16, pad=20)
ax3.zaxis.set_major_formatter(FormatStrFormatter('%.1f%%'))
ax3.view_init(elev=30, azim=-120)

# Add legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=plt.cm.viridis(0.5), edgecolor='k', label='Rough (alpha < 1)'),
    Patch(facecolor=plt.cm.plasma(0.5), edgecolor='k', label='Classical (alpha = 1)')
]
ax3.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(0.8, 0.95))

output_file_3 = 'results/iv_surface_comparison.png'
fig3.savefig(output_file_3, dpi=300)
print(f"Plot 3 saved to '{output_file_3}'")


# Finally, display all figure windows
plt.show()
```

```python
#!/usr/bin/env python3
"""
Utility for extracting ATM implied volatility skew from the Monte Carlo IV surface
and (optionally) plotting a three-panel figure.

The IV surface generator (src/cdq_mc_iv.cpp) writes CSV rows with columns:
  Maturity, LogMoneyness, IV_Rough, IV_Classical

This script groups rows by maturity, finds the centre point (log-moneyness  0),
and performs a small-window linear regression to estimate the slope of the IV
curve around ATM. By default the regression uses points with |log-moneyness|  0.04.

Outputs one CSV line per maturity:
  Maturity,ATM_IV,Skew_dLogMoneyness,Skew_dStrike

Skew_dLogMoneyness  = /log(K/S0) evaluated at ATM.
Skew_dStrike        = /K at K = S0, using the relation /K = (1/K) /log(K).

Extra:
  If --plot is provided, the script will also produce a 1x3 panel figure:
  (a) ATM skew vs maturity, (b) ATM IV vs maturity,
  (c) power-law fit |skew|  c  T^{-} on log-log axes, with  and R annotated.

Example
-------
    python scripts/compute_atm_skew.py \
        --csv results/iv_surface_cdq.csv \
        --model rough \
        --spot 100.0 \
        --plot rough_vol_three_panel.png \
        --skew dlogm
"""

from __future__ import annotations

import argparse
import csv
import sys
from collections import defaultdict
from pathlib import Path
from typing import Dict, Iterable, List, Tuple


def _read_surface_rows(path: Path) -> Iterable[Dict[str, float]]:
    """Yield parsed rows from the IV surface CSV, skipping banner lines."""
    with path.open(newline="") as fh:
        reader = csv.reader(fh)
        header: List[str] | None = None
        idx_map: Dict[str, int] | None = None
        for row in reader:
            if not row:
                continue
            if header is None:
                if row[0].strip().startswith("Maturity"):
                    header = [col.strip() for col in row]
                    idx_map = {name: header.index(name) for name in header}
                continue
            if len(row) < len(header):
                continue
            assert header is not None and idx_map is not None  # for type checkers
            record = {
                "Maturity": float(row[idx_map["Maturity"]]),
                "LogMoneyness": float(row[idx_map["LogMoneyness"]]),
                "IV_Rough": float(row[idx_map["IV_Rough"]]),
                "IV_Classical": float(row[idx_map["IV_Classical"]]),
            }
            yield record


def _linear_regression_slope(points: List[Tuple[float, float]]) -> float:
    """Return dY/dX for points via least squares; raises if X variance is tiny."""
    n = len(points)
    if n < 2:
        raise ValueError("Need at least two points to estimate a slope.")
    sum_x = sum(x for x, _ in points)
    sum_y = sum(y for _, y in points)
    mean_x = sum_x / n
    mean_y = sum_y / n
    cov = sum((x - mean_x) * (y - mean_y) for x, y in points)
    var_x = sum((x - mean_x) ** 2 for x, _ in points)
    if var_x <= 1e-18:
        raise ValueError("Variance of x too small for regression.")
    return cov / var_x


def compute_atm_skew(
    rows: Iterable[Dict[str, float]],
    model_column: str,
    window: float,
    spot: float,
) -> List[Tuple[float, float, float, float]]:
    """Compute ATM IV and skew for each maturity."""
    by_maturity: Dict[float, List[Tuple[float, float]]] = defaultdict(list)
    for row in rows:
        by_maturity[row["Maturity"]].append((row["LogMoneyness"], row[model_column]))

    results: List[Tuple[float, float, float, float]] = []
    for maturity in sorted(by_maturity):
        points = sorted(by_maturity[maturity], key=lambda kv: kv[0])
        if not points:
            continue
        centre = min(points, key=lambda kv: abs(kv[0]))
        local_pts = [(k, iv) for k, iv in points if abs(k) <= window + 1e-12]
        if len(local_pts) < 2:
            raise RuntimeError(
                f"Not enough points near ATM for maturity {maturity:g}; "
                f"try increasing --window."
            )
        slope_log = _linear_regression_slope(local_pts)
        atm_iv = centre[1]
        # /K = (1/K) /log(K) with K = S0 * exp(m), m=0  K = S0.
        slope_strike = slope_log / spot
        results.append((maturity, atm_iv, slope_log, slope_strike))
    return results


# -------------------- Plotting helpers --------------------
def _power_law_fit(T: List[float], skew_vals: List[float]):
    """Fit |skew|  c * T^{-beta} in log-space. Return (c, beta, r2)."""
    import numpy as np

    T_arr = np.asarray(T, dtype=float)
    y_arr = np.abs(np.asarray(skew_vals, dtype=float))

    mask = (T_arr > 0.0) & (y_arr > 0.0)
    if not np.any(mask):
        raise ValueError("No positive data to fit power law.")
    x = np.log(T_arr[mask])
    y = np.log(y_arr[mask])

    # y = a + b x  |skew| = exp(a) * T^b; define beta = -b
    b, a = np.polyfit(x, y, 1)  # slope, intercept
    y_hat = a + b * x
    ss_res = float(np.sum((y - y_hat) ** 2))
    ss_tot = float(np.sum((y - y.mean()) ** 2))
    r2 = 1.0 - ss_res / ss_tot if ss_tot > 0 else 1.0
    c = float(np.exp(a))
    beta = float(-b)
    return c, beta, r2


def _plot_three_panel(
    results: List[Tuple[float, float, float, float]],
    skew_kind: str,
    out_path: Path,
    show: bool,
) -> None:
    """Make a 1x3 panel: (a) skew vs T, (b) IV vs T, (c) power-law fit on log-log axes."""
    import numpy as np
    import matplotlib.pyplot as plt

    T, iv, skew_log, skew_dK = zip(*results)
    T = np.asarray(T, float)
    iv = np.asarray(iv, float)
    skew_vals = np.asarray(skew_log if skew_kind == "dlogm" else skew_dK, float)

    # Power-law fit parameters on |skew|
    c, beta, r2 = _power_law_fit(T.tolist(), skew_vals.tolist())
    T_fit = np.linspace(T.min(), T.max(), 200)
    skew_fit = c * (T_fit ** (-beta))

    plt.figure(figsize=(14, 4.2))

    # (a) skew vs maturity
    ax1 = plt.subplot(1, 3, 1)
    ax1.plot(T, skew_vals, marker="o")
    ax1.set_title("ATM Skew vs Maturity")
    ax1.set_xlabel("Maturity (years)")
    if skew_kind == "dlogm":
        ax1.set_ylabel("ATM Skew (d / d log K)")
    else:
        ax1.set_ylabel("ATM Skew (d / dK)")
    ax1.grid(True, alpha=0.4)

    # (b) IV vs maturity
    ax2 = plt.subplot(1, 3, 2)
    ax2.plot(T, iv, marker="o")
    ax2.set_title("ATM IV vs Maturity")
    ax2.set_xlabel("Maturity (years)")
    ax2.set_ylabel("ATM implied volatility")
    ax2.grid(True, alpha=0.4)

    # (c) power-law on log-log axes
    ax3 = plt.subplot(1, 3, 3)
    ax3.set_title("Power-law fit (log scale)")
    ax3.set_xlabel("Maturity (years) [log]")
    ax3.set_ylabel("|ATM skew| [log]")
    ax3.set_xscale("log")
    ax3.set_yscale("log")
    ax3.plot(T, np.abs(skew_vals), marker="o", linestyle="None")
    ax3.plot(T_fit, np.abs(skew_fit), linewidth=1.5)
    ax3.grid(True, which="both", alpha=0.4)
    ax3.text(
        0.05,
        0.95,
        fr"|skew|  cT^(-)    = {beta:.3f},  R = {r2:.3f}",
        transform=ax3.transAxes,
        va="top",
    )

    # (a)(b)(c) tags
    for ax, tag in zip([ax1, ax2, ax3], ["(a)", "(b)", "(c)"]):
        ax.text(0.02, -0.2, tag, transform=ax.transAxes, fontsize=12)

    plt.tight_layout()
    out_path = Path(out_path)
    out_path.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    if show:
        plt.show()
    plt.close()


# -------------------- CLI --------------------
def main() -> None:
    parser = argparse.ArgumentParser(description="Compute ATM IV skew from the Monte Carlo surface.")
    parser.add_argument(
        "--csv",
        type=Path,
        default=Path("results/iv_surface_cdq.csv"),
        help="Path to the IV surface CSV generated by cdq_mc_iv.cpp.",
    )
    parser.add_argument(
        "--model",
        choices=("rough", "classical"),
        default="rough",
        help="Select which IV column to analyse.",
    )
    parser.add_argument(
        "--window",
        type=float,
        default=0.04,
        help="Half-width (in log-moneyness) of the regression window around ATM.",
    )
    parser.add_argument(
        "--spot",
        type=float,
        default=100.0,
        help="Spot level used to convert d/dlogK into d/dK.",
    )
    # New: output & plotting options
    parser.add_argument(
        "--save-csv",
        type=Path,
        default=None,
        help="Optional path to save the per-maturity CSV; if omitted, write to stdout.",
    )
    parser.add_argument(
        "--plot",
        type=Path,
        default=None,
        help="If provided, save a 1x3 panel plot to this path (e.g., rough_vol_three_panel.png).",
    )
    parser.add_argument(
        "--skew",
        choices=("dlogm", "dK"),
        default="dlogm",
        help="Which skew to plot/fit: dlogm = /logK (default), dK = /K.",
    )
    parser.add_argument(
        "--show",
        action="store_true",
        help="Show the plot window in addition to saving it when --plot is given.",
    )

    args, _ = parser.parse_known_args()

    column_map = {
        "rough": "IV_Rough",
        "classical": "IV_Classical",
    }

    rows = list(_read_surface_rows(args.csv))
    if not rows:
        raise SystemExit(f"No data rows found in {args.csv}")

    results = compute_atm_skew(rows, column_map[args.model], args.window, args.spot)

    def _write(rows_to_write: List[Tuple[float, float, float, float]], fh):
        writer = csv.writer(fh)
        writer.writerow(["Maturity", "ATM_IV", "Skew_dLogMoneyness", "Skew_dStrike"])
        for maturity, atm_iv, slope_log, slope_strike in rows_to_write:
            writer.writerow(
                [
                    f"{maturity:.6f}",
                    f"{atm_iv:.6f}",
                    f"{slope_log:.6f}",
                    f"{slope_strike:.6f}",
                ]
            )

    save_csv_path = args.save_csv
    if save_csv_path is not None:
        with save_csv_path.open("w", newline="") as fh:
            _write(results, fh)
    else:
        _write(results, sys.stdout)

    # Optional plotting
    if args.plot is not None:
        try:
            _plot_three_panel(results, args.skew, args.plot, args.show)
        except ImportError as e:
            raise SystemExit(
                "Plotting requires numpy and matplotlib. Please install them, e.g.: "
                "`pip install numpy matplotlib`"
            ) from e


if __name__ == "__main__":
    try:
        main()
    except Exception as exc:  # pragma: no cover - CLI safety net
        print(f"Error: {exc}", file=sys.stderr)
        raise SystemExit(1) from exc
```

```python
import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ======= Data: paste the provided table directly =======
csv = """Maturity,ATM_IV,Skew_dLogMoneyness,Skew_dStrike
0.083333,0.207432,-0.084705,-0.000847
0.166667,0.212290,-0.071260,-0.000713
0.250000,0.216134,-0.061590,-0.000616
0.333333,0.219978,-0.038935,-0.000389
0.416667,0.222845,-0.044590,-0.000446
0.500000,0.225271,-0.040025,-0.000400
0.583333,0.227141,-0.046015,-0.000460
0.666667,0.229533,-0.039000,-0.000390
0.750000,0.231776,-0.045675,-0.000457
0.833333,0.233936,-0.037325,-0.000373
0.916667,0.235538,-0.032785,-0.000328
1.000000,0.237648,-0.032365,-0.000324
"""
df = pd.read_csv(io.StringIO(csv.strip()))

# If you prefer d/dK for the skew, switch to the line below
# skew_col = "Skew_dStrike"
skew_col = "Skew_dLogMoneyness"

T = df["Maturity"].to_numpy()
IV = df["ATM_IV"].to_numpy()
skew = df[skew_col].to_numpy()

# ======= Compute the power-law fit |skew| = c * T^{-beta} =======
# Run OLS in log space; natural log or log10 both produce the same slope
x = np.log(T)
y = np.log(np.abs(skew))

# Fit: y = a + b * x  =>  |skew| = exp(a) * T^b
b, a = np.polyfit(x, y, 1)
beta = -b                   # because we write it as |skew| = c * T^{-beta}
c = np.exp(a)
y_hat = a + b * x
ss_res = np.sum((y - y_hat)**2)
ss_tot = np.sum((y - y.mean())**2)
r2 = 1 - ss_res/ss_tot

# Plot the fitted curve back on the original (non-log) scale
T_fit = np.linspace(T.min(), T.max(), 200)
skew_fit = c * (T_fit ** (-beta))

# ======= Plotting (1x3 layout matching the example figure)=======
plt.figure(figsize=(14, 4.2))

# (a) ATM skew vs. maturity
ax1 = plt.subplot(1, 3, 1)
ax1.plot(T, skew, marker="o")
ax1.set_title("ATM Skew vs Maturity")
ax1.set_xlabel("Maturity (years)")
ylabel = "ATM Skew (d / d log K)" if skew_col == "Skew_dLogMoneyness" else "ATM Skew (d / dK)"
ax1.set_ylabel(ylabel)
ax1.grid(True, alpha=0.4)

# (b) ATM IV vs. maturity
ax2 = plt.subplot(1, 3, 2)
ax2.plot(T, IV, marker="o")
ax2.set_title("ATM IV vs Maturity")
ax2.set_xlabel("Maturity (years)")
ax2.set_ylabel("ATM implied volatility")
ax2.grid(True, alpha=0.4)

# (c) Power-law fit (log axes)
ax3 = plt.subplot(1, 3, 3)
ax3.set_title("Power-law fit (log scale)")
ax3.set_xlabel("Maturity (years) [log]")
ax3.set_ylabel("|ATM skew| [log]")
ax3.set_xscale("log")
ax3.set_yscale("log")
ax3.plot(T, np.abs(skew), marker="o", linestyle="None")
ax3.plot(T_fit, np.abs(skew_fit), linewidth=1.5)
ax3.grid(True, which="both", alpha=0.4)

# Annotate the parameters in the upper-right corner
txt = fr"|skew|  cT^(-)    = {beta:.3f},   R = {r2:.3f}"
ax3.text(0.05, 0.95, txt, transform=ax3.transAxes, va="top")

# Lowercase subplot labels
for ax, tag in zip([ax1, ax2, ax3], ["(a)", "(b)", "(c)"]):
    ax.text(0.02, -0.2, tag, transform=ax.transAxes, fontsize=12)

plt.tight_layout()
plt.savefig("rough_vol_three_panel.png", dpi=300, bbox_inches="tight")
plt.show()

print(f"Fitted power law: |skew|  {c:.6g} * T^(-{beta:.3f});  R^2 = {r2:.3f}")
```
